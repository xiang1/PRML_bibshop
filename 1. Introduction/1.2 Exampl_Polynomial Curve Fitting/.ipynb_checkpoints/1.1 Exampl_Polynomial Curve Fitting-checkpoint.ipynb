# Example: Polynomial Curve Fitting

采用多项式模型区拟合未知模型的数据
### 1.未知的数据模型 ： $ t = \sin \left( 2\pi x\right)+ \epsilon $

   其中观测数据 $t$ 带有噪声$\epsilon$ ,噪声满足一定的分布
 

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline
N = 10;
x_observe = np.linspace(0, 1, N);
t_observe = np.sin(2 * np.pi * x_observe) + 0.2 * np.random.randn(N);

x_real = np.linspace(0,1,200);
t_real = np.sin(2 * np.pi * x_real);

fig,ax = plt.subplots();
ax.plot(x_observe, t_observe,'co',label = 'observe data');
ax.plot(x_real, t_real, 'g',label = 'real model data');
plt.legend(loc = 'upper right');
plt.xlabel("$x $", fontsize = 'x-large'); plt.ylabel('$t $', fontsize = 'x-large');
```

我们做机器学习，或者说模式识别的目的在于，从观测数据中学习到其中的模型，当出现新的观测自变量$\hat{x} $时，能够很好的预测他的输出 $ \hat{t} $

### 2. 多项式模型： $y \left( x, \bf{w} \right) = \omega_0 + \omega_1 x + \omega_2 x^2 +... \omega_M x^M = \sum_{j = 0}^M \omega_j x^j$

   其中$\bf{w}$ 被称为多项式的阶数， 虽然模型是自变量$x$的非线性函数，但是由于模型关于参数 $\bf{w}$ 是线性的，所以这种类型的模型被称为线性模型，为了求其中的参数$\bf{w}$ ，需要通过拟合原始的数据来达到，这就需要我们来定义一个误差函数来衡量拟合效果，最常用的误差函数是用误差的平方和的形式。
    
### 3. 误差函数 ：$E \left(\bf{w}\right) = \frac{1}{2} \sum_{n = 1}^{N} \{ y\left(x_n,\bf{w}\right)- t_n \}^2$ 

   其中的参数 $\frac{1}{2}$是为了分析的方便，以后会解释；（**注： 后面由于需要误差函数的最小化，需要令导数为0，二次函数求导之后会出来参数2，刚好消掉**）
   令 $\bf{w} = \left( \omega_0, \omega_1,\ldots, \omega_M \right)^{T}$, $\bf{t} = \left(t_1, t_2, \dots, t_N \right)^{T}$, 
   
   $\bf{x} =\begin{bmatrix}   
            x_1^0 & \cdots & x_1^M\\ 
           \cdots & x_n^m  & \cdots \\ 
            x_N^0 & \cdots & x_N^M
        \end{bmatrix} $
        
   $$ 
   E \left(\bf{w}\right) = \frac{1}{2} \left( \bf{x}\bf{w}-\bf{t} \right)^{T} \left( \bf{x}\bf{w}-\bf{t} \right)\\
   \left. \frac{\partial E \left(\bf{w}\right)}{ \partial \bf{w}}\right|_{\bf{w} = \bf{w}_*} = \bf{w}_*^T\bf{x}^T\bf{x}-\bf{t}^T\bf{x}=0\\
   $$
   因此
   $$
   \bf{w_*} =\left[\bf{x}^T\bf{x}\right]^{-1}\bf{x}^T\bf{t} 
   $$

### 4. 研究不同多项式阶数对于拟合效果的影响
   

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline
N = 10;
x_observe = np.linspace(0, 1, N);
t_observe = np.sin(2 * np.pi * x_observe) + 0.2 * np.random.randn(N);

x_real = np.linspace(0,1,200);
t_real = np.sin(2 * np.pi * x_real);
plt.figure(figsize=(12,8));

M = 0;
plt.subplot(2,2,1);
coeff_0 = np.polyfit(x_observe,t_observe, M);
f_0 = np.poly1d( coeff_0 );
t_polymodel_0 = f_0(x_real);
plt.plot(x_observe, t_observe,'co', x_real, t_real, 'g-', x_real, t_polymodel_0, 'r');
plt.xlabel(r'$x$'); plt.ylabel(r'$t$'); plt.text(0.6,1, r'$M = 0$');

M = 1;
plt.subplot(2,2,2);
coeff_1 = np.polyfit(x_observe,t_observe, M);
f_1 = np.poly1d( coeff_1 );
t_polymodel_1 = f_1(x_real);
plt.plot(x_observe, t_observe,'co', x_real, t_real, 'g-', x_real, t_polymodel_1, 'r');
plt.xlabel(r'$x$'); plt.ylabel(r'$t$'); plt.text(0.6,1, r'$M = 1$');

M = 3;
plt.subplot(2,2,3);
coeff_3 = np.polyfit(x_observe,t_observe, M);
f_3 = np.poly1d( coeff_3 );
t_polymodel_3 = f_3(x_real);
plt.plot(x_observe, t_observe,'co', x_real, t_real, 'g-', x_real, t_polymodel_3, 'r');
plt.xlabel(r'$x$'); plt.ylabel(r'$t$'); plt.text(0.6,1, r'$M = 3$');

M = 9;
plt.subplot(2,2,4);
coeff_9 = np.polyfit(x_observe,t_observe, M);
f_9 = np.poly1d( coeff_9 );
t_polymodel_9 = f_9(x_real);
plt.plot(x_observe, t_observe,'co', x_real, t_real, 'g-', x_real, t_polymodel_9, 'r');
plt.xlabel(r'$x$'); plt.ylabel(r'$t$'); plt.text(0.6,1, r'$M = 9$');
plt.show();
""" 从图中可以看出，M= 0 和 M= 1 时，学习到的模型和真实模型相去比较远，
    而 M= 3 时候和真实的模型最为接近，
    M =9 时虽然曲线过了所有的观测点，但是出现了过拟合的现象（也就是和原始模型并不相近）"""
```

### 5. 多项式的阶数 M 的选取
   
   通常，为了保持误差和原始函数的尺度和单位保持一致，用另外一种方法来衡量误差，也就是均方误差 ： $E_{RMS} \sqrt{2E\left(\bf{w}_*\right)/N}$
   
   不管训练数据的规模有多大，都可以保持一样的尺度。在测试数据中用$E_{RMS}$来衡量模型泛化能力。

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline
N = 10;
x_tr = np.linspace(0, 1, N);   t_tr = np.sin(2 * np.pi * x_tr) + 0.2 * np.random.randn(10);
x_te = np.linspace(0, 1, 100); t_te =  np.sin(2 * np.pi * x_te) + 0.2 * np.random.randn(100);
# 注： 这里出现了问题，其中的rand()缺省的话为1，也就是说对于前面变量中的每个数据都会产生一样的随机数，

rms_tr = []; rms_te = []; #误差矩阵
for M in range(10):
    coeff = np.polyfit(x_tr, t_tr, M);
    f = np.poly1d(coeff);
    rms_tr_tmp = np.sqrt( ((f(x_tr)-t_tr)**2).sum()/x_tr.shape[0] );
    rms_tr.append(rms_tr_tmp);
    rms_te_tmp = np.sqrt( ((f(x_te)-t_te)**2).sum()/x_te.shape[0] ); 
    rms_te.append(rms_te_tmp);
    
plt.plot(range(10), rms_tr,'bo-', range(10), rms_te, 'ro-');
plt.xlabel('M');  plt.ylabel('$E_{RMS }$');
plt.legend(['training data', 'test data'], loc = "best");
plt.show();
""" 
当阶数比较大的时候，模型为了和训练数据更加贴近，使得参数变得非常大，也就是说模型曲线波动非常大
"""
print(coeff)
```

### 6. 训练数据集大小和过拟合之间的关系

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline
N = 15;
x_observe = np.linspace(0, 1, N);
t_observe = np.sin(2 * np.pi * x_observe) + 0.2 * np.random.randn(N);

x_real = np.linspace(0,1,200);
t_real = np.sin(2 * np.pi * x_real);
plt.figure(figsize=(12,4));

M = 9;
plt.subplot(1,2,1);
coeff_0 = np.polyfit(x_observe,t_observe, M);
f_0 = np.poly1d( coeff_0 );
t_polymodel_0 = f_0(x_real);
plt.plot(x_observe, t_observe,'bo', x_real, t_real, 'g-', x_real, t_polymodel_0, 'r');
plt.xlabel(r'$x$'); plt.ylabel(r'$t$'); 
plt.legend(['observe data ', 'real date','fit with M =9, N = 15']);
## 训练数据集为100
N = 100;
x_observe =np.linspace(0, 1, N);
t_observe = np.sin(2 * np.pi * x_observe) + 0.2 * np.random.randn(N);

M = 9;
plt.subplot(1,2,2);
coeff_0 = np.polyfit(x_observe,t_observe, M);
f_0 = np.poly1d( coeff_0 );
t_polymodel_0 = f_0(x_real);
plt.plot(x_observe, t_observe,'bo', x_real, t_real, 'g-', x_real, t_polymodel_0, 'r');
plt.xlabel(r'$x$'); plt.ylabel(r'$t$');
plt.legend(['observe data ', 'real date','fit with M =9, N = 100']);
```

** 数据集大小的增大能够降低过拟合程度 **

另外一种说法也就是说，当数据集越大，我们能够承受越复杂的模型。一个比较粗略的说法便是，我们的数据集大小应该至少是模型中可变参数的5~10倍，但是我们也需要明确一点的是，模型复杂度并不是单纯用模型中可变参数的个数来衡量。

最小二乘方法表示的是最大似然方法的一种特殊情况（maxinun likelihood）（** 注： 假设观测误差呈现高斯分布**）， 过拟合现象是最大似然方法的一个比较一般的特性，如果采用贝叶斯方法，过拟合现象能够被有效的避免。

### 7. 正则化方法

避免上述过拟合现象发生的一个有效技巧是采用正则化方法，也就是在误差函数中增加一个正则化项，正则化项的目的是希望能够限制模型中可变参数的规模，避免参数过大。
$$
   \tilde{E}\left(\bf{w}\right) = \frac{1}{2}\sum_{n = 1}^{N} \{ y\left(x_n,\bf{w}\right)- t_n \}^2 + \frac{\lambda}{2}\left\| \bf{w}\right\|
$$
其中， $\left\| \bf{w}\right\| = \bf{w}^T\bf{w} = \omega_0^2 +\omega_1^2 +\ldots + \omega_M^2$ (**注：通常模型中的参数$\omega_0$ 不被放入正则化项中，因为，$\omega_0 $ 通常取决于目标变量的中心化方法，后面讨论 **)， 二次正则化项被称为 redge regression. 

通过最小化的$\tilde{E}\left(\bf{w}\right)$，可以得到：
$$
\tilde{E} \left(\bf{w}\right) = \frac{1}{2} \left( \bf{x}\bf{w}-\bf{t} \right)^{T} \left( \bf{x}\bf{w}-\bf{t} \right) +\frac{\lambda}{2}\bf{w}^T\bf{w}\\
   \left. \frac{\partial \tilde{E} \left(\bf{w}\right)}{ \partial \bf{w}}\right|_{\bf{w} = \bf{w}_*} = \bf{w}_*^T\bf{x}^T\bf{x}-\bf{t}^T\bf{x} + \lambda \bf{w_*}=0\\
$$
因此可以计算最优的参数$\bf{w}_*$: $ \bf{w}_* = \left[\bf{x}^T\bf{x}+ \lambda \bf{I}\right]^{-1}\bf{x}^T\bf{t} $


```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt


N = 10;
x_observe = np.linspace(0, 1, N);
t_observe = np.sin(2 * np.pi * x_observe) + 0.2 * np.random.randn(N);
x_real = np.linspace(0, 1, 200);
t_real = np.sin(2 * np.pi * x_real);

M = 9; 
lambd =0;

x_matrix = np.vstack([x_observe**0, x_observe**1, x_observe**2, x_observe**3, x_observe**4, x_observe**5,x_observe**6,\
                     x_observe**7, x_observe**8, x_observe**9]).transpose();
x_real_matrix = np.vstack([x_real**0, x_real**1, x_real**2, x_real**3, x_real**4, x_real**5,x_real**6,\
                     x_real**7, x_real**8, x_real**9]);
x_real_matrix = x_real_matrix.transpose().copy();
matrix_tmp1= x_matrix.transpose().copy();                 #矩阵转置 x^T
matrix_tmp2 = np.dot(matrix_tmp1, x_matrix).copy();       # 矩阵相乘x^T * x ，点乘用 a*b
matrix_tmp3 = matrix_tmp2 + lambd*np.eye(M+1);            # 单位矩阵 x^T * x+ \lambda* I
matrix_tmp4 = np.linalg.inv(matrix_tmp3);                 # 矩阵求逆 (x^T * x+ \lambda* I)^(-1)
matrix_tmp5 = np.dot(matrix_tmp4, x_matrix.transpose()).copy();
matrix_tmp5 = np.dot(matrix_tmp5, t_observe).copy();

plt.figure(figsize=(12,4));
plt.subplot(1,2,1);
coeff = matrix_tmp5.copy();
t_predict = np.dot(x_real_matrix,coeff).copy();
plt.plot(x_observe, t_observe,'bo',x_real, t_real, 'c-', x_real, t_predict,'r-');
plt.xlabel('x'); plt.ylabel('y'); plt.legend(['observe data', 'real model','learned data with $\lambda$ = 0']);

## lambda = 1;
lambd = 1;
matrix_tmp3 = matrix_tmp2 + lambd*np.eye(M+1);            # 单位矩阵 x^T * x+ \lambda* I
matrix_tmp4 = np.linalg.inv(matrix_tmp3);                 # 矩阵求逆 (x^T * x+ \lambda* I)^(-1)
matrix_tmp5 = np.dot(matrix_tmp4, x_matrix.transpose()).copy();
matrix_tmp5 = np.dot(matrix_tmp5, t_observe).copy();

plt.subplot(1,2,2);
coeff = matrix_tmp5.copy();
t_predict = np.dot(x_real_matrix,coeff ).copy();
plt.plot(x_observe, t_observe,'bo',x_real, t_real, 'c-', x_real, t_predict,'r-');
plt.xlabel('x'); plt.ylabel('y'); plt.legend(['observe data', 'real model','learned data with $\lambda$ = 1']);
```

### 9. 正则化参数的选取


```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

%matplotlib inline
N = 11;
x_tr = np.linspace(0, 1, N);   t_tr = np.sin(2 * np.pi * x_tr) + 0.2 * np.random.randn(N);
x_te = np.linspace(0, 1, 100); t_te =  np.sin(2 * np.pi * x_te) + 0.2 * np.random.randn(100);
# 注： 这里出现了问题，其中的rand()缺省的话为1，也就是说对于前面变量中的每个数据都会产生一样的随机数，
x_matrix = np.vstack([x_tr**0, x_tr**1, x_tr**2, x_tr**3, x_tr**4, x_tr**5,x_tr**6,\
                     x_tr**7, x_tr**8, x_tr**9]).transpose();
x_real_matrix = np.vstack([x_te**0, x_te**1, x_te**2, x_te**3, x_te**4, x_te**5,x_te**6,\
                     x_te**7, x_te**8, x_te**9]).transpose();
rms_tr = []; rms_te = []; #误差矩阵
for lambd in range(-36, 1, 1): 
    matrix_tmp1= x_matrix.transpose().copy();                 #矩阵转置 x^T
    matrix_tmp2 = np.dot(matrix_tmp1, x_matrix).copy();       # 矩阵相乘x^T * x ，点乘用 a*b
    lambd0 = np.exp(lambd);
    matrix_tmp3 = matrix_tmp2 + lambd0*np.eye(M+1);            # 单位矩阵 x^T * x+ \lambda* I
    matrix_tmp4 = np.linalg.inv(matrix_tmp3);                 # 矩阵求逆 (x^T * x+ \lambda* I)^(-1)
    matrix_tmp5 = np.dot(matrix_tmp4, x_matrix.transpose()).copy();
    matrix_tmp6 = np.dot(matrix_tmp5, t_tr).copy();
    coeff = matrix_tmp6.copy();
    t_predict_observe = np.dot(x_matrix,coeff).copy();
    t_predict = np.dot(x_real_matrix, coeff).copy();    
    rms_tr_tmp = np.sqrt( ((t_predict_observe-t_tr)**2).sum()/x_tr.shape[0] );
    rms_tr.append(rms_tr_tmp);
    rms_te_tmp = np.sqrt( ((t_predict-t_te)**2).sum()/x_te.shape[0] ); 
    rms_te.append(rms_te_tmp);
    
plt.plot(range(-36, 1, 1), rms_tr,'b-', range(-36, 1, 1), rms_te, 'r-');
plt.xlabel(' $\ln \lambda $');  plt.ylabel('$E_{RMS }$');
plt.legend(['training', 'test'], loc = "best");
```

随着正则化项系数的增大，训练误差逐渐增加，而测试的误差先减小后又出现增大的情况，第一阶段，一个增加，一个减小，意味着模型的过拟合情况降低，模型与真实模型趋于接近，在第二阶段，两个误差均增大，意味着选择的模型完全不能反应真实的情况。
